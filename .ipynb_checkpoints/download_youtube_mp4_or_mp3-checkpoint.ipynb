```python
from pytube import YouTube
from moviepy.editor import AudioFileClip
import os

def download_video(url, format_choice):
    try:
        yt = YouTube(url)
        
        if format_choice == 'mp4':
            # Download video in mp4 format
            video_stream = yt.streams.filter(file_extension="mp4", progressive=True).get_highest_resolution()
            print(f"Downloading {yt.title} in MP4 format...")
            video_stream.download()
            print("Download completed!")

        elif format_choice == 'mp3':
            # Download audio and convert to mp3
            video_stream = yt.streams.filter(only_audio=True).first()
            print(f"Downloading {yt.title} in MP3 format...")
            download_path = video_stream.download(filename="temp_audio")
            
            # Convert to MP3
            audio = AudioFileClip(download_path)
            audio.write_audiofile(yt.title + ".mp3")
            audio.close()
            
            # Remove the temporary audio file
            os.remove(download_path)
            print("Download completed as MP3!")
        
        else:
            print("Invalid format choice. Choose 'mp3' or 'mp4'.")
    except Exception as e:
        print("Error:", e)

# Input from user
url = input("Enter YouTube video URL: ")
format_choice = input("Enter format (mp3 or mp4): ").lower()
download_video(url, format_choice)
```
